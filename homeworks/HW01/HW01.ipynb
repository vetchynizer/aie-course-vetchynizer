```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4, 5]) # одномерный массив из обычного списка
type(a) # отображение типа массива (обычный n-dimensional array)
```

```python
a.dtype # отображение типа элементов массива (int)
```

```python
a.shape # отображение формы массива (лишь одно число элементов, ведь массив одномерный)
```

```python
b = np.array([[1, 2, 3], [3, 4, 5]]) # двумерный массив
b
```

```python
b.shape # форма двумерного массива
```

```python
zero=np.zeros((2, 3)) # двумерный массив с нулями
zero
```

```python
one=np.ones((2, 3)) # двумерный массив с единицами
one
```

```python
c=np.arange(5, 15, 2) # создание массива с указанием начала, конца и шага (конец не включительно)
c
```

```python
c[3] # получение третьего элемента массива (первый элемент считаем нулевым)
```

```python
c[1:4] # срез с первого по четвертый элемент не включительно
```

```python
c[[1,4]] # срез с выбором первого и четвертого элементов
```

```python
sum = a+c # поэлементное сложение (только для массивов одинаковой формы)
sum
```

```python
mul = a*c # поэлементное умножение
mul
```

```python
power = c**a # возведение в степень
power
```

```python
remain = c%a # остаток от деления
remain
```

```python
a + 1 # сложение со скаляром
```

```python
a * 2 # умножение на скаляр
```

```python
c.min(), c.max() # минимум и максимум из массива c
```

```python
b.mean() # средняя величина всего массива
```

```python
b.mean(axis=0) # средняя величина по столбцам
```

```python
b.mean(axis=1) # средняя величина по строкам
```
